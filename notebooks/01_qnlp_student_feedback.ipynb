# QNLP: Student Feedback (starter)

This notebook demonstrates loading the sample dataset and running the classical baseline.
If you have `lambeq` and `pennylane` installed you can extend the `src/feedback_qnlp/model.py` quantum stub.

```python
from feedback_qnlp import load_data, preprocess_texts, vectorize_texts, train_classical_baseline
df = load_data('data/student_feedback.csv')
texts = preprocess_texts(df['feedback'].astype(str).tolist())
X, vec = vectorize_texts(texts)
y = (df['label'] == 'positive').astype(int).tolist()
result = train_classical_baseline(X, y)
print('Classical baseline accuracy:', result['accuracy'])
```
